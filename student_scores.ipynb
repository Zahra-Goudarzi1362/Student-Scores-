# Predict of scores of student base on Hours, Practice & TeamWork 
# Through Machine Learning model 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression 
from sklearn import metrics
```

```python
data = pd.read_csv ('E:\Momeni-Machne Learning class momeni 21 01 1402\Machine Learning\dataset/student_scores.csv')
data
```

```python
df = pd.DataFrame (data, columns=['Hours','Practice','TeamWork','Scores'])
df
```

```python
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
x= pd.DataFrame(data, columns=['Hours','Practice','TeamWork'])
y = data['Scores'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
regressor = LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred = regressor.predict(x_test)
```

```python
a= x_test.Hours
b= y_test
c= x_test.Hours
d= y_pred
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.scatter (a,b)
plt.scatter(c,d)
plt.show()
```

```python
a= x_test.Hours
b= y_test
c= x_test.Hours
d= y_pred
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.scatter (a,b)
plt.plot(c,d,color='red')
plt.show()
```
